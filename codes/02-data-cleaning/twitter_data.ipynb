```python
import pandas as pd
import time
import numpy as np
import re
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
df = pd.read_csv('../01-data-collection/twitter_scrape.csv',index_col=0)
df.head(3)
```

```python
df_en = df[df['lang']=='en']
df_en = df_en[~df_en.text.str.contains('RT')].copy()
df_en.shape
```

```python
df_en.drop('geo',axis=1,inplace=True)
```

```python
df_en
```

```python
df_en.text
```

```python
#1.clean the unrelated words
lemmatizer = WordNetLemmatizer()
stopwords = stopwords.words('English')
```

```python
def pipeline(input:str):
    input = input.lower()
    input=re.sub("\\W"," ",input) # remove special chars
    str_input = re.sub(pattern=r'[^a-zA-z]',repl=' ',string=input)
#     str_output = correct_sentence_spelling(sentence = str_input) # it is too time-consuming
    token_list = word_tokenize(str_input)
    token_list = [lemmatizer.lemmatize(i) for i in token_list]
    word_list = [word for word in token_list if word not in stopwords]
    word_str = ' '.join(word_list)
    return word_str
```

```python
df_en.text = df_en.text.apply(pipeline)
```
